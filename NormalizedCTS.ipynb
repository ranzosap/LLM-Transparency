```python
import matplotlib.pyplot as plt
import numpy as np

# Labels and angle configuration
labels = ['Code', 'Weights', 'Data', 'Docs', 'License', 'Carbon', 'Benchmarks']
num_vars = len(labels)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]

# Normalized CTS components (all binary  normalized to [0,1])
all_models = {
    'ChatGPT-4': [0, 0, 0, 1, 0, 0, 0],
    'DeepSeek-R1': [1, 1, 0, 0, 0, 1, 0],
    'BLOOM': [1, 1, 1, 1, 1, 1, 1],
    'Llama 2': [1, 1, 0, 0, 0, 0, 0],
    'Llemma 34B': [1, 1, 0, 1, 1, 0, 0],
    'Mistral 7B': [1, 1, 0, 0, 1, 0, 0],
    'Gemini 1.5': [0, 0, 0, 1, 0, 0, 0],
    'T5-XXL': [1, 1, 1, 1, 1, 0, 0],
    'Grok-3': [0, 0, 0, 1, 0, 0, 0],
    'CodeLlama': [1, 1, 0, 0, 0, 0, 0],
    'GPT-3.5 Turbo': [0, 0, 0, 1, 0, 0, 0],
    'PaLM 2': [1, 1, 0, 1, 1, 0, 0],
    'Claude 2': [0, 0, 0, 1, 0, 0, 0],
    'Claude 3 Opus': [0, 0, 0, 1, 0, 0, 0],
    'OPT': [1, 1, 1, 1, 1, 0, 1],
    'Falcon 40B': [1, 1, 1, 1, 1, 0, 0],
    'Vicuna-13B': [1, 1, 0, 1, 1, 0, 0],
    'Orca 2': [1, 1, 0, 1, 0, 0, 0],
    'Alpaca-7B': [1, 1, 1, 1, 1, 0, 0],
    'WizardLM-2': [1, 1, 0, 1, 1, 0, 0],
}

# Divide models into groups: 3 groups with 3 models, rest with 2 or 1
model_groups = [
    ['ChatGPT-4', 'DeepSeek-R1', 'BLOOM'],
    ['Llama 2', 'Llemma 34B', 'Mistral 7B'],
    ['Gemini 1.5', 'T5-XXL', 'Grok-3'],
    ['CodeLlama', 'GPT-3.5 Turbo'],
    ['PaLM 2', 'Claude 2'],
    ['Claude 3 Opus', 'OPT'],
    ['Falcon 40B', 'Vicuna-13B'],
    ['Orca 2', 'Alpaca-7B'],
    ['WizardLM-2']
]

# Plot each group
for i, group in enumerate(model_groups):
    fig, ax = plt.subplots(figsize=(9, 9), subplot_kw=dict(polar=True))
    colors = plt.cm.get_cmap('tab10', len(group))

    for idx, model in enumerate(group):
        values = all_models[model] + all_models[model][:1]
        ax.plot(angles, values, label=model, color=colors(idx), linewidth=3)
        ax.fill(angles, values, color=colors(idx), alpha=0.15)

    ax.set_theta_offset(np.pi / 2)
    ax.set_theta_direction(-1)
    ax.set_ylim(0, 1)

    ax.set_thetagrids(np.degrees(angles[:-1]), labels, fontsize=14, fontweight='bold')
    ax.set_rgrids([0, 0.5, 1.0], labels=['0', '0.5', '1.0'], angle=0, fontsize=12)

    plt.title(f'Normalized CTS Components by Model Group {i+1}', fontsize=18, fontweight='bold', pad=20)
    ax.legend(loc='lower center', bbox_to_anchor=(0.5, -0.25), ncol=2, fontsize=12, frameon=False)
    plt.tight_layout()
    plt.show()
```
