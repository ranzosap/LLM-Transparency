```python
import matplotlib.pyplot as plt
import numpy as np

labels = ['Code', 'Weights', 'Data', 'Docs', 'License', 'Carbon', 'Benchmarks']
num_vars = len(labels)

models = {
    'ChatGPT-4': [0, 0, 0, 1, 0, 0, 0],
    'DeepSeek-R1': [1, 1, 0, 0, 0, 1, 0],
    'BLOOM': [1, 1, 1, 1, 1, 1, 1],
    'Llama 2': [1, 1, 0, 0, 0, 0, 0],
    'Llemma 34B': [1, 1, 0, 1, 1, 0, 0],
    'Mistral 7B': [1, 1, 0, 0, 1, 0, 0],
    'Gemini 1.5': [0, 0, 0, 1, 0, 0, 0],
    'T5-XXL': [1, 1, 1, 1, 1, 0, 0],
    'Grok-3': [0, 0, 0, 1, 0, 0, 0],
    'CodeLlama': [1, 1, 0, 0, 0, 0, 0],
}

angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]

fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(polar=True))

colors = plt.cm.get_cmap('tab10', len(models))
for idx, (name, scores) in enumerate(models.items()):
    values = scores + scores[:1]
    ax.plot(angles, values, label=name, color=colors(idx), linewidth=4)
    ax.fill(angles, values, color=colors(idx), alpha=0.08)

ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)
ax.set_ylim(0, 1)

# Set bold, large, black axis (category) labels
ax.set_thetagrids(
    np.degrees(angles[:-1]),
    labels,
    fontsize=26,
    fontweight='bold',
    color='black'
)

# Set bold, large, black radial grid labels
ax.set_rgrids(
    [0, 1],
    labels=['0', '1'],
    angle=0,
    fontsize=26,
    fontweight='bold',
    color='black'
)

# Set bold, large, black title
plt.title(
    'Composite Transparency Score (CTS) Components by Model',
    fontsize=36,
    fontweight='bold',
    color='black',
    pad=40
)

plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1.05), fontsize='small')
plt.tight_layout()
plt.show()
```
